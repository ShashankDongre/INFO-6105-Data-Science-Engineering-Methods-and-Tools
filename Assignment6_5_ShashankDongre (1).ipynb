```python
class ProbDist(dict):
    """A Probability Distribution; an {outcome: probability} mapping."""
    def __init__(self, mapping=(), **kwargs):
        self.update(mapping, **kwargs)
        # Make probabilities sum to 1.0; assert no negative probabilities
        total = sum(self.values())
        for outcome in self:
            self[outcome] = self[outcome] / total
            assert self[outcome] >= 0
```

```python
from fractions import Fraction
def p(event, space): 
    """The probability of an event, given a sample space of equiprobable outcomes. 
    event: a collection of outcomes, or a predicate that is true of outcomes in the event. 
    space: a set of outcomes or a probability distribution of {outcome: frequency} pairs."""
    # branch on the type of the first argument
    if is_predicate(event):
        # transform the mapping (untangible) 'event' into the collection (tangible) 'event'
        event = such_that(event, space)
        
    if isinstance(space, ProbDist):
        # if space is a dictionary of distinct probabilities, where each item does not count as the same amount
        return sum(space[o] for o in space if o in event)
    else:
        # space is not a dictionary but a collection, let's fall back to our original division
        return Fraction(len(event & space), len(space))

is_predicate = callable

def such_that(predicate, space): 
    """The outcomes in the sample pace for which the predicate is true.
    If space is a set, return a subset {outcome,...} with outcomes where predicate(element) is true;
    if space is a ProbDist, return a ProbDist {outcome: frequency,...} with outcomes where predicate(element) is true."""
    if isinstance(space, ProbDist):
        return ProbDist({o:space[o] for o in space if predicate(o)})
    else:
        return {o for o in space if predicate(o)}
```

### What is the Probability Distribution for each F1 driver to win both the Bahrain and Saudi Arabian Grand Prix?

```python
# Since these are the latest driver standings available on the website and considering the fact that Saudi Arabia GP is just one week after the
# Bahrain GP, We assume that the two races are independent events i.e. the outcome of Bahrain GP does not affect the outcome of Saudi Arabia GP. 
# And we also consider the above driver standings and their probability distribution as our initial data and We also assume the probability 
# distribution of each driver remains the same.

# To get the Probability Distribution for each F1 driver to win both the Bahrain and Saudi Arabian Grand Prix, we multiply the two probabilties 
# to arrive at the conclusion of BGP and SAGP.  
# We take these two dictionaries from the result obtained above and store the resulting probability distribution in a dictionary named both_wins 

both_wins = {}
for key in BGP:
    if key in SAGP:
        both_wins[key] = BGP[key] * SAGP[key]
print(both_wins)
```

This means that 'MV' has a 3.7% chance of winning both races, 'CL' has 1.7% chance of winning both races and so on...

```python
# We now define the universal sample space BOTH as a joint distribution of Probability Distribution for each F1 driver to win both 
# the Bahrain and Saudi Arabian Grand Prix. 
# Outcome 'MV MV' denotes the probability of 'MV' winning the first race and second race as well. 'MV CL' denotes the probability of 'MV' winning
# the first race and 'CL' winning the seccond race, and so on...Here, we will use set comprehension to get the universal sample space named as BOTH

def joint(A, B, sep=''):
    """The joint distribution of two independent probability distributions. 
    Result is all entries of the form {a+sep+b: P(a)*P(b)}"""
    return ProbDist({a + sep + b: A[a] * B[b] 
                        for a in A 
                        for b in B})

BOTH = joint(BGP, SAGP, ' ')
BOTH
```

### If Ferrari wins the first race, what is the probability that Ferrari wins the next one?

```python
# Now for ferrari to win the second race after winning the first, there are only four combinations possible 'CL CS', 'CS CL', 'CS CS', and 'CL CL'
# We now isolate them as below and get the required probability for given the condition: If Ferrari wins the first race, 
# what is the probability that Ferrari wins the next one? 

def ferrari_win_second_after_winning_first(outcome):
    return outcome.startswith('CS CS') or outcome.startswith('CL CL') or outcome.startswith('CL CS') or outcome.startswith('CS CL')
p(ferrari_win_second_after_winning_first, such_that(ferrari_win_first, BOTH))
```

###  If Ferrari wins at least one of these two races, what is the probability Ferrari wins both races?

```python
# If Ferrari wins the first race, what is the probability that Ferrari wins the next one?
# The first probability is p(ferrari_win_first, BOTH) calculates the probability that a Ferrari wins the first race given the sample space BOTH.
# The second probability is p(ferrari_win_second, BOTH) calculates the probability that a Ferrari wins the second race given the sample space BOTH.
# The last p function calculates the conditional probability for Ferrari to win the second race given the prior event that Ferrari wins the first race.

def ferrari_win_first(outcome):
    return outcome.startswith('CS') or outcome.startswith('CL')
p(ferrari_win_first, BOTH)
def ferrari_win_second(outcome):
    return outcome.endswith('CS') or outcome.endswith('CL')
p(ferrari_win_second, BOTH)

p(ferrari_win_second, such_that(ferrari_win_first, BOTH))
```

Hence, the conditional probability for Ferrari to win the second race given the prior event that Ferrari wins the first race is 0.2357446808510639 or 23.57%

### Ferrari wins one of these two races on a rainy day. What is the probability Ferrari wins both races, assuming races can be held on either rainy, sunny, cloudy, snowy or foggy days? Assume that rain, sun, clouds, snow, and fog are the only possible weather conditions on race tracks and there is an equal probability for each one of these weather events

```python
# Assume that rain, sun, clouds, snow, and fog are the only possible weather conditions on race tracks and there is an equal probability for 
# each one of these weather events, which makes this an equi-probable event. Hence, we can use set comprehension.
# weather denotes the Probability Distribution of each weather condition.

weather = ProbDist(
    rain=0.20,
     sun=0.20,
  clouds=0.20,
    snow=0.20,
     fog=0.20)

weather
```

```python
# We now define the a sample space JBGP as a joint distribution of BGP and weather. 
# Joining BGP and weather

def joint(A, B, sep=''):
    """The joint distribution of two independent probability distributions. 
    Result is all entries of the form {a+sep+b: P(a)*P(b)}"""
    return ProbDist({a + sep + b: A[a] * B[b]
                    for a in A
                    for b in B})

JBGP = joint(BGP, weather, ' ')
JBGP
```

```python
# We now define the a sample space JSAGP as a joint distribution of JSAGP and weather. 
# Joining JSAGP and weather

def joint(A, B, sep=''):
    """The joint distribution of two independent probability distributions. 
    Result is all entries of the form {a+sep+b: P(a)*P(b)}"""
    return ProbDist({a + sep + b: A[a] * B[b]
                    for a in A
                    for b in B})

JSAGP = joint(SAGP, weather, ' ')
JSAGP
```

```python
# We now define the a joint sample space which would be our universal sample space named J_total as a joint distribution of JBGP and JSAGP. 
# Joining JBGP and JSAGP

def joint(A, B, sep=''):
    """The joint distribution of two independent probability distributions. 
    Result is all entries of the form {a+sep+b: P(a)*P(b)}"""
    return ProbDist({a + sep + b: A[a] * B[b]
                    for a in A
                    for b in B})

J_total = joint(JBGP, JSAGP, ' ')
J_total
```

### What is the probability Ferrari wins both races, assuming races can be held on either rainy, sunny, cloudy, snowy or foggy days?

```python
# The ferrari_wins_atleast_one_race_on_rainy_days function takes a single argument outcome, 
# which is a string. The function returns True if either the string 'CL rain' or 
# the string 'CS rain' is contained in outcome and False otherwise.
# The p function calculates the probability that Ferrari wins at least one race on rainy days, 
# given the joint distribution of the outcomes J_total.
# We now calculate the probability of Ferrari wins one of these two races on a rainy day.

def ferrari_wins_atleast_one_race_on_rainy_days(outcome):
    return 'CL rain' in outcome or 'CS rain' in outcome

p(ferrari_wins_atleast_one_race_on_rainy_days, J_total)
```

```python
# The function both_races_ferrari takes a string argument outcome and returns a boolean value indicating whether Ferrari won both races.
# We use .split(' ') method to split the outcome string into a list of substrings separated by spaces, 
# and then indexing is used to retrieve the first and third elements of the list and assign them to win_BGP_ferrari and win_SAGP_ferrari respectively.
# The function then returns True if either the string 'CL' is contained in win_BGP_ferrari and the string 'CS' is contained in win_SAGP_ferrari 
# or the string 'CS' is contained in win_BGP_ferrari and the string 'CL' is contained in win_SAGP_ferrari. 
# Thus the probability of Ferrari winning both races, given the prior event that Ferrari wins at least one race on a rainy day is computed.

def both_races_ferrari(outcome):
    win_BGP_ferrari = outcome.split(' ')[0]
    win_SAGP_ferrari = outcome.split(' ')[2]
    
    return (('CL' in win_BGP_ferrari and 'CS' in win_SAGP_ferrari) or
            ('CS' in win_BGP_ferrari and 'CL' in win_SAGP_ferrari) or
            ('CL' in win_BGP_ferrari and 'CL' in win_SAGP_ferrari) or
            ('CS' in win_BGP_ferrari and 'CS' in win_SAGP_ferrari)) 

p(both_races_ferrari, such_that(ferrari_wins_atleast_one_race_on_rainy_days, J_total))
```

Thus the probability of Ferrari winning both races, given the prior event that Ferrari wins one of the two races on a rainy day 
is 0.21729277433975333 or 21.72%
